```python
import itertools
```

```python
# hns cyl project:
a_cs = [1, 2]
bend = [2]  # bending rigidity of monomer-monomer interactions.

n_mon = [200]
r_cyl = [4.5]
l_cyl = [75]
n_hns = [0, 4, 8, 12, 16, 20, 30]
rand_g_s = list(range(70000,70000+len(n_hns)*1000,1000))  # random group for a space
phi_c = [0, 0.08, 0.12, 0.16, 0.2, 0.24, 0.28, 0.32, 0.36, 0.4]
rand_g_e = [g+i*10 for g in rand_g_s for i in range(len(phi_c))]  # random group for a ensemble
```

```python
list(zip(n_hns, rand_g_s))
```

```python
ensembles = []
for a_c  in a_cs:
    for n_h, r_g_s in list(zip(n_hns, rand_g_s)):
        rand_g_e = [r_g_s+i*10 for i in range(len(phi_c))]
        if a_c == 1:
            ensembles.extend(list(itertools.product(
                bend, n_mon, r_cyl, l_cyl, [n_h], rand_g_e, [a_c], phi_c
            )))
        else: 
            ensembles.extend(list(itertools.product(
                bend, n_mon, r_cyl, l_cyl, [n_h], rand_g_e, [a_c], phi_c[1:]
            )))
```
